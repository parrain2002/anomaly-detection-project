```python
# 03_Comparaison_Models.ipynb

# üìå Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# üì• Chargement des pr√©dictions
# Remplace par tes chemins ou variables selon ton organisation
isoforest_preds = pd.read_csv("../results/isoforest_predictions.csv")
svm_preds = pd.read_csv("../results/oneclasssvm_predictions.csv")

# üßÆ Affichage de quelques infos
print("üîç Aper√ßu des r√©sultats Isolation Forest")
print(isoforest_preds["prediction"].value_counts())
print("\nüîç Aper√ßu des r√©sultats One-Class SVM")
print(svm_preds["prediction"].value_counts())

# üìä Matrice de confusion
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues")
    plt.title(title)
    plt.xlabel("Pr√©dit")
    plt.ylabel("R√©el")
    plt.show()

# üß© Affichage des matrices
plot_confusion_matrix(isoforest_preds["true_label"], isoforest_preds["prediction"], "Matrice - Isolation Forest")
plot_confusion_matrix(svm_preds["true_label"], svm_preds["prediction"], "Matrice - One-Class SVM")

# üìã Rapports de classification
print("\nüìà Rapport Isolation Forest")
print(classification_report(isoforest_preds["true_label"], isoforest_preds["prediction"]))

print("\nüìà Rapport One-Class SVM")
print(classification_report(svm_preds["true_label"], svm_preds["prediction"]))
```
